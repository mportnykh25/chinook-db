```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('Chinook_Sqlite.sqlite')
df = pd.read_sql("""
    SELECT g.Name as Genre, COUNT(t.TrackId) as Tracks
    FROM Track t
    JOIN Genre g ON t.GenreId = g.GenreId
    GROUP BY g.Name
    ORDER BY Tracks DESC
    LIMIT 5
""", conn)
conn.close()

print(df)
```
